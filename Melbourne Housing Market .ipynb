## Filling in the missing values

#### Bedroom2, Bathroom(/s) and Car(/s) are the columns where the missing values should be filled because these are the most important features with respect to the price of the house and the number of values missing are less than 30%.

* Postal Code, Regionname,Propertycount and Distance have only one value missing so we can drop that column.
* CouncilArea, YearBuilt, BuildingArea, Landsize will be difficult to fill in. They either have too many values missing or are difficult features to fill in without adequate information.



##### Filling in the Bedroom2 values

```python
df['Bedroom2'].unique()
```

```python
df['Bedroom2']=df['Bedroom2'].transform(lambda x:x.fillna(x.median()))
```

```python
df['Bedroom2'].isnull().any()
```

```python
print(df['Bathroom'].nunique())
print(df['Bathroom'].unique())
```
